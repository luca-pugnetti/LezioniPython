# **Esercizio 1 - Esercizio Lezione 9 per casa**

Crea una lista di tuple dove ogni tupla rappresenta un prodotto (nome, prezzo, quantità). /
Implementa le seguenti funzioni:
- Aggiungi prodotto
- Rimuovi prodotto
- Aggiorna quantità
- Calcola valore totale

```python
prodotti = []

def aggiungi_prodotto(nome, prezzo, quantita):
  for i, elemento in enumerate(prodotti):
    if elemento[0] == nome.lower():
      print("Il prodotto è già presente nella lista")
      return

  prodotti.append((nome.lower(), prezzo, quantita))
  print("Il prodotto è stato aggiunto con successo")

def rimuovi_prodotto(nome):
  for i, elemento in enumerate(prodotti):
    if elemento[0] == nome.lower():
      prodotti.pop(i)
      print("Il prodotto è stato rimosso con successo")
      return

  print("Il prodotto è inesistente")

def aggiorna_quantita(nome, nuova_quantita):
  for i, elemento in enumerate(prodotti):
    if elemento[0] == nome.lower():
      prodotti[i] = (prodotti[i][0], prodotti[i][1], nuova_quantita)
      print("La quantità è stata aggiornata con successo")
      return

  print("Il prodotto è inesistente")

def calcola_valore_totale():
  valore_totale = 0

  for prodotto in prodotti:

    valore_totale = valore_totale + prodotto[1]*prodotto[2]

  return valore_totale

aggiungi_prodotto("Mela", 0.5, 10)
aggiungi_prodotto("Pera", 0.4, 8)
rimuovi_prodotto("Ananas")
rimuovi_prodotto("mela")
aggiorna_quantita("pera", 6)
valore_totale = calcola_valore_totale()

print("Prodotti: ", prodotti)
print("Valore totale: ", valore_totale)
```

# **Esercizio 1 - Esercizio Lezione 10 in classe**

Si consideri una lista di dizionari, ciascuno contenente le informazioni di uno studente. Ogni dizionario include i seguenti campi:

- nome: il nome dello studente (stringa),
- eta: l'età dello studente (intero),
- voto: il voto finale dello studente (intero).

`studenti = [
    {'nome': 'Alice', 'eta': 20, 'voto': 75},
    {'nome': 'Bob', 'eta': 17, 'voto': 45},
    {'nome': 'Carla', 'eta': 22, 'voto': 88},
    {'nome': 'Daniele', 'eta': 19, 'voto': 55},
    {'nome': 'Eva', 'eta': 16, 'voto': 65},
    {'nome': 'Francesco', 'eta': 21, 'voto': 90}
]`

L'obiettivo dell'esercizio è di eseguire le seguenti operazioni:

- Creazione della lista di dizionari contenente le informazioni sugli studenti.
- Creazione di una lista di tuple che contenga il nome e il voto degli studenti che hanno superato l'esame, ovvero quelli con un voto maggiore di 50.
- Filtraggio degli studenti maggiorenni, cioè degli studenti con età pari o superiore a 18 anni, utilizzando la funzione filter() e una lambda.
Creazione di un dizionario che raggruppi gli studenti in base alla fascia di voto, con le seguenti categorie:
  - "pass" per i voti compresi tra 50 e 69,
  - "good" per i voti compresi tra 70 e 84,
  - "excellent" per i voti pari o superiori a 85.

Stampa dei risultati:
La lista degli studenti che hanno superato l'esame (voto maggiore di 50),
La lista degli studenti maggiorenni, senza alcun ordinamento, nell'ordine in cui appaiono nella lista originale,
Le fasce di voto degli studenti, con il nome e il voto di ciascuno.

```python
studenti = [ {'nome': 'Alice', 'eta': 20, 'voto': 75},
             {'nome': 'Bob', 'eta': 17, 'voto': 45},
             {'nome': 'Carla', 'eta': 22, 'voto': 88},
             {'nome': 'Daniele', 'eta': 19, 'voto': 55},
             {'nome': 'Eva', 'eta': 16, 'voto': 65},
             {'nome': 'Francesco', 'eta': 21, 'voto': 90} ]

# Creazione della lista di tuple (nome, voto) per gli studenti con voto > 50
studenti_passati = [(studente['nome'], studente['voto']) for studente in studenti if studente['voto'] > 50]

# Filtraggio studenti maggiorenni
studenti_adulti = filter(lambda studente: studente['eta'] >= 18, studenti)
studenti_adulti = list(studenti_adulti)

# Creazione di un dizionario che raggruppa gli studenti per fasce di voto
fasce_voto = {
    'pass': [studente for studente in studenti if studente['voto'] >= 50 and studente['voto'] < 70],
    'good': [studente for studente in studenti if studente['voto'] >= 70 and studente['voto'] < 85],
    'excellent': [studente for studente in studenti if studente['voto'] >= 85]
}


print("\nStudenti che hanno passato l'esame")
for nome, voto in studenti_passati:
  print(f"{nome}: {voto}")

print("\nStudenti maggiorenni")
for studente in studenti_adulti:
  print(f"{studente['nome']}: {studente['eta']}")

print("\nFasce di voto degli studenti")
for fascia, lista in fasce_voto.items():
  print(f"{fascia.capitalize()}")
  for studente in lista:
    print(f" {studente['nome']} - Voto: {studente['voto']}")
```
