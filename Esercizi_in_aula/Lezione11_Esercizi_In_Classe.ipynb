# **Esercizio 1**

Crea un dizionario che rappresenti gli studenti e le loro materie scelte.
- Studente: Marco ha scelto "Matematica" e "Fisica".
- Studente: Luca ha scelto "Matematica".
- Studente: Sofia ha scelto "Fisica" e "Chimica".

Aggiungi un nuovo studente:
- Anna con le materie "Informatica" e "Fisica".

Modifica le materie di uno studente:
- Lo studente Luca aggiunge la materia "Chimica".

Rimuovi una materia:
- Lo studente Marco abbandona la materia "Fisica".

Confronta le materie scelte dagli studenti con le materie offerte:
- Materie offerte: "Matematica", "Fisica", "Chimica", "Informatica", "Biologia".

Trova:
- L'elenco delle materie scelte almeno una volta.
- L'elenco delle materie offerte che nessuno ha scelto.

Stampa un riepilogo con:
- Elenco degli studenti e delle loro materie.
- Materie scelte almeno una volta.
- Materie offerte ma non scelte.
- Eventuali discrepanze tra le materie scelte dagli studenti e quelle offerte.

```python
# Dizionario iniziale degli studenti
studenti = {
    "Marco": {"Matematica", "Fisica"},
    "Luca": {"Matematica"},
    "Sofia": {"Fisica", "Chimica"},
    "Giovanni": {"Storia"}
}

# Aggiunta di Anna con informatica e fisica
studenti['Anna'] = {"Informatica", "Fisica"}

# Luca aggiunge chimica
studenti['Luca'].add("Chimica")

# Marco rimuove fisica
studenti['Marco'].remove("Fisica")

# Materie offerte
materie_offerte = frozenset({"Matematica", "Fisica", "Chimica", "Informatica", "Biologia"})

# Riepilogo delle materie scelte e non scelte
materie_scelte = set.union(*studenti.values())
materie_non_scelte = materie_offerte - materie_scelte

# Elenco studenti e materie
for studente, materie in studenti.items():
  print(f"Studente: {studente}, Materie: {materie}")

# Materie scelte almeno una volta e non scelte
print("Materie scelte almeno una volta: ", materie_scelte)
print("Materie non scelte: ", materie_non_scelte)

# Controllo discrepanze
for studente, materie in studenti.items():

  extra_materie = materie - materie_offerte
  if extra_materie:
    print(f"Lo studente {studente} ha scelto materie non offerte {extra_materie}")
```
