# **Esercizio 1**
Un mazzo di carte contiene quattro semi (cuori, quadri, fiori, picche) e valori da 1 a 13. Ogni giocatore pesca casualmente una carta dal mazzo e il vincitore viene determinato in base alle seguenti regole:



1. Se la carta di un giocatore ha un valore maggiore dell'altro, quel giocatore vince.
2. Se le carte hanno lo stesso valore:

 1. Vince il giocatore con il seme "cuori".

 2. Se nessuno dei due ha cuori, vince chi ha "quadri".

 3. Se entrambi hanno lo stesso seme, il gioco termina in pareggio.

Scrivi un programma che:
- Utilizzi una funzione per generare casualmente una carta per ogni giocatore.

- Determini il vincitore usando condizioni e condizioni nidificate.
- Mostri le carte pescate e il vincitore.

**Nota**: Il mazzo di carte viene rappresentato con un valore da 1 a 13 per il valore e una stringa ("cuori", "quadri", "fiori", "picche") per il seme. I semi hanno prioritÃ  in quest'ordine: cuori > quadri > fiori > picche.

```python
import random

seme = ["picche", "fiori", "quadri", "cuori"]

valore1 = random.randint(1, 13)
indice_seme1 = random.randint(0, 3)
print("Giocatore 1: ", valore1, seme[indice_seme1])

valore2 = random.randint(1, 13)
indice_seme2 = random.randint(0, 3)
print("Giocatore 2: ", valore2, seme[indice_seme2])

giocatore1_vince = False
giocatore2_vince = False

if valore1 > valore2:

  giocatore1_vince = True

elif valore1 < valore2:

  giocatore2_vince = True

else:

    if indice_seme1 > indice_seme2:

      giocatore1_vince = True

    elif indice_seme1 < indice_seme2:

      giocatore2_vince = True


if giocatore1_vince:

  print("Ha vinto il giocatore 1")

elif giocatore2_vince:

  print("Ha vinto il giocatore 2")

else:

  print("Pareggio")
```
