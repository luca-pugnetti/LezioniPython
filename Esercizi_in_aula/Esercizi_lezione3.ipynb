```python
import random

def verifica_moltiplicazione():
    numero1 = random.randint(1, 10)
    numero2 = random.randint(1, 10)
    risposta = int(input(f"Quanto vale {numero1} * {numero2}? "))

    risposta_corretta = numero1*numero2

    if risposta == risposta_corretta:
      print("Molto bene!")

    else:
      print("No. Prova ancora.")


verifica_moltiplicazione()
```

```python
import random

def verifica_moltiplicazione():

    livello = int(input(f"Scegli il livello di difficoltà: 1, 2 o 3? "))

    if livello == 1:

      limite_inferiore = 1
      limite_superiore = 9

    elif livello == 2:

      limite_inferiore = 10
      limite_superiore = 99

    else:

      limite_inferiore = 100
      limite_superiore = 999

    numero1 = random.randint(limite_inferiore, limite_superiore)
    numero2 = random.randint(limite_inferiore, limite_superiore)
    risposta = int(input(f"Quanto vale {numero1} * {numero2}? "))

    risposta_corretta = numero1*numero2

    if risposta == risposta_corretta:
      print("Molto bene!")

    else:
      print("No. Prova ancora.")


verifica_moltiplicazione()
```

```python
def fahrenheit(gradi_celsius):

  gradi_fahr = (9/5) * gradi_celsius +32

  return gradi_fahr

fahrenheit(30)
```

```python
def mistery(x):

  y=0

  for value in x:

    y = y + value ** 2

  return y

print(mistery([1,2,3,4,5]))
```
