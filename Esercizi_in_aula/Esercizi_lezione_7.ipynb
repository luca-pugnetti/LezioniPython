# **Esercizio 1**

Scrivi una funzione ricorsiva chiamata `somma_cifre` che prenda un numero intero positivo come input e restituisca la somma delle sue cifre.  Ad esempio:

* `somma_cifre(123)` dovrebbe restituire 6 (1 + 2 + 3)
* `somma_cifre(9)` dovrebbe restituire 9
* `somma_cifre(12345)` dovrebbe restituire 15
* `somma_cifre(0)` dovrebbe restituire 0

```python
def somma_cifre(n):

  if n < 10:
    return n

  else:
    return n % 10 + somma_cifre(n//10)

print(somma_cifre(123))
print(somma_cifre(9))
print(somma_cifre(12345))
```

# **Esercizio 2**

Scrivi una funzione ricorsiva chiamata `calcola_potenza` che prenda due numeri interi positivi, base ed esponente, come input e restituisca il risultato di base elevato alla potenza di esponente.

Ad esempio:

- `calcola_potenza(2, 3)` dovrebbe restituire 8 ($2^3 = 2 * 2 * 2 = 8$)
- `calcola_potenza(5, 2)` dovrebbe restituire 25 ($5^2 = 5 * 5 = 25$)
- `calcola_potenza(10, 0)` dovrebbe restituire 1 (qualsiasi numero elevato alla potenza di 0 è 1)
- `calcola_potenza(7, 1)` dovrebbe restituire 7 (qualsiasi numero elevato alla potenza di 1 è il numero stesso)

```python
def calcola_potenza(base, esponente):

  print(base, esponente)

  if esponente == 0:
    return 1

  elif esponente == 1:
    return base

  else:
    return base * calcola_potenza(base, esponente-1)

print(calcola_potenza(2, 3))
print("\n")
print(calcola_potenza(10, 1))
```

# **Esercizio 1**
Creare una lista di n numeri interi, quindi, al termine dell’inserimento, sommare gli elementi pari e dispari separatamente. Visualizzare in output le due somme ottenute.



```python
quantita_numeri = int(input("Quanti elementi vuoi inserire? "))

somma_pari = 0
somma_dispari = 0

numeri_inseriti = []

for i in range(quantita_numeri):

    numero_inserito = int(input("Inserisci un numero: "))
    numeri_inseriti.append(numero_inserito)

for i in range(quantita_numeri):

    if numeri_inseriti[i]%2 == 0:

      somma_pari+=numeri_inseriti[i]

    else:

      somma_dispari+=numeri_inseriti[i]

print(somma_pari, somma_dispari)
```

# **Esercizio 2**
Popolare una lista di n elementi con i primi n multipli di 10. I numeri devono essere inseriti in coda alla lista.
Dopo l’inserimento visualizzare in output i valori della lista e il relativo indice.
Poi modificare solo gli elementi maggiori di 30 sottraendo la metà del numero. Visualizzare nuovamente la lista.


```python
quantita_numeri = int(input("Quanti multipli di 10 conterrà la lista? "))

lista = []

moltiplicatore = 10

for i in range(quantita_numeri):

  lista.append(i*moltiplicatore)

for i in range(quantita_numeri):

  print("Indice: ", i, "Elemento: ", lista[i])

for i in range(quantita_numeri):

  if lista[i] > 30:

    lista[i] = lista[i]//2

for i in range(quantita_numeri):

  print("Indice: ", i, "Elemento: ", lista[i])
```
