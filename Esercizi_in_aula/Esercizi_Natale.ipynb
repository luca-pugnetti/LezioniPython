# Esercizio 1

Scrivere un programma che chieda all'utente di inserire una data nel formato e tramite una funzione verifichi se quella data corrisponde al giorno di Natale. Date ammesse:
- 25 dicembre
- 25/12
- 25 Dicembre

Il programma dovrebbe stampare un messaggio che indica se la data è Natale o meno.

```python
def stringa_natale(stringa):

  return stringa in ["25 dicembre", "25/12", "25 Dicembre"]

data = input("Inserisci una data: ")

risultato = stringa_natale(data)

if risultato:

  print("Natale")

else:

  print("Non è Natale")
```

## Esercizio 2
Scrivi una funzione ricorsiva che disegna un albero di Natale. Il programma deve stampare l'albero con caratteri * in modo che l'albero abbia un numero crescente di stelle ad ogni livello, fino a raggiungere un massimo di 5 livelli.

image.png

```python
def disegna_albero(livello):

  if livello == 0:
    return

  disegna_albero(livello-1)

  print(livello)

  print(" " * (5 - livello) + "*" * (2 * livello - 1))

  if livello == 5:

    print(" " * ((2 * livello - 1 )//2) + '|')

disegna_albero(5)
```

# Esercizio 3

Scrivi una funzione che prende una lista di regali (ad esempio, una lista di stringhe come ["puzzle", "giocattolo", "libro"]) e la ordina in ordine alfabetico, poi la stampa.

```python
def ordina_regali(regali):

  ordinati = []
  while regali:

    minimo = regali[0]

    for regalo in regali:
      if regalo < minimo:
        minimo = regalo

    ordinati.append(minimo)
    regali.remove(minimo)

  return ordinati

regali = ["puzzle", "giocattolo", "libro", "bambola", "macchina"]

regali_ordinati = ordina_regali(regali.copy())

print(regali, regali_ordinati)
```

```python
regali = ["puzzle", "giocattolo", "libro", "bambola", "macchina"]

regali_copia = regali.copy()

regali_copia.sort()

print(regali, regali_copia)
```

# Esercizio 4

Scrivi una funzione che prende una lista di desideri natalizi (una lista di stringhe) e consente all'utente di aggiungere, rimuovere e visualizzare i desideri. Il programma deve continuare a eseguire finché l'utente non decide di uscire.

```python
def gestisci_lista(lista_desideri):

  while True:

    print("\nCosa vuoi fare?")
    print("1. Aggiungi un desiderio")
    print("2. Rimuovi un desiderio")
    print("3. Visualizza la lista dei desideri")
    print("0. Esci")

    scelta = input("Inserisci il numero dell'opzione: ")

    if scelta == "1":

      desiderio = input("Inserisci un desiderio")
      lista_desideri.append(desiderio)

    elif scelta == "2":

      if not lista_desideri:
        print("Lista dei desideri vuota")
        continue

      print("\nLista dei desideri: ")

      for i in range(len(lista_desideri)):

        print("Indice: ", i, ", Desiderio: ", lista_desideri[i])

      indice_da_rimuovere = int(input("Inserisci l'indice del desiderio da rimuovere: "))

      if 0 <= indice_da_rimuovere < len(lista_desideri):

        desiderio_rimosso = lista_desideri.pop(indice_da_rimuovere)
        print("Desiderio ", desiderio_rimosso, "rimosso con successo!")

      else:

        print("Indice non valido")
        continue

    elif scelta == "3":

      if not lista_desideri:

        print("\nLa lista dei desideri è vuota")

      else:

        print("\nLista dei desideri:")

        for desiderio in lista_desideri:

          print(desiderio)

    elif scelta == "0":

      print("Uscita dal programma....")
      break

    else:

      print("Operazione non valida!!")

gestisci_lista([])
```
