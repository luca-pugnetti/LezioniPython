```python
# orario di uscita in secondi
orario_uscita = (6 * 3600) + (52 * 60)

# ritmi in secondi
ritmo_blando = 8 * 60 + 15
ritmo_moderato = 7 * 60 + 12

# tempo totale di corsa
tempo_totale_corsa = (2 * ritmo_blando) + (3 * ritmo_moderato)

# orario di rientro in secondi
orario_rientro = orario_uscita + tempo_totale_corsa

# conversione in ore e minuti

# siccome 1 ora = 3600 sec --> la divisione intera dell'orario di rientro (in secondi) per 3600 ci dar√† il numero di ore
ore_rientro = orario_rientro // 3600

# ora analizziamo il resto della divisione per 3600 e ci dar√† la parte di ora non ancora utilizzata e dividendola (tramite
# divisione intera) per 60 otteniamo il numero di minuti
minuti_rientro = (orario_rientro % 3600) // 60

print("Orario rientro: {:02}:{:02}".format(ore_rientro, minuti_rientro))
```
